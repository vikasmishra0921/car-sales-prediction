```python
cars_data['mileage'] = cars_data['mileage'].str.extract(r'(\d+\.\d+|\d+)')
cars_data['mileage'] = cars_data['mileage'].astype(float)



cars_data
```

```python
cars_data['engine'] = cars_data['engine'].str.extract(r'(\d+\.\d+|\d+)').astype(float)
cars_data['max_power'] = cars_data['max_power'].str.extract(r'(\d+\.\d+|\d+)').astype(float)
```

```python
cars_data['seats'].fillna(cars_data['seats'].mode()[0], inplace=True)
```

```python
# For continuous features (use median to avoid outlier impact)
cars_data['mileage'].fillna(cars_data['mileage'].median(), inplace=True)
cars_data['engine'].fillna(cars_data['engine'].median(), inplace=True)
cars_data['max_power'].fillna(cars_data['max_power'].median(), inplace=True)

# For seats (discrete values, so use mode)
cars_data['seats'].fillna(cars_data['seats'].mode()[0], inplace=True)
```

```python
def get_brand_name(car_name):
    car_name = car_name.split(' ')[0]
    return car_name.strip()
```

```python
cars_data['name'] = cars_data['name'].apply(get_brand_name)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
cars_data['name'] = le.fit_transform(cars_data['name'])  # converts brand names to numbers
```

```python
cars_data.reset_index(inplace=True)
```

```python
cars_data['owner'].unique()
```

```python

cars_data['owner'] = le.fit_transform(cars_data['owner'])  
```

```python
cars_data.drop(columns=['index'], inplace=True)
```

```python
X= cars_data.drop(columns=['selling_price'])
y =cars_data['selling_price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor 

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Define your models
models = {
    'LinearRegression': LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'DecisionTree': DecisionTreeRegressor(),
    'RandomForest': RandomForestRegressor(),
    'gradient' : GradientBoostingRegressor()
    
}

# Dictionary to hold scores
results = {}

# Loop through models and evaluate
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    results[name] = {
        'R2 Score': round(r2, 4),
        'MAE': round(mae, 2),
        'MSE': round(mse, 2),
        'RMSE': round(rmse, 2)
    }

# Print the results
print("Model Performance Comparison:\n")
for model_name, scores in results.items():
    print(f"{model_name} ")
    for metric, value in scores.items():
        print(f"   {metric}: {value}")
    print('-' * 40)
```
